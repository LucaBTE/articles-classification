```python
import numpy as np
import pandas as pd

DATA_PATH = 'bbc_articles.csv'
df = pd.read_csv(DATA_PATH, sep=';')

print(f'Dataset shape: {df.shape}')
display(df.head(3))
print('\nClass distribution:')
df = df[df['category'] == 'sport']
print(df['category'].value_counts())
```

```python
# Combine title + text for a richer input
df['title'] = df['title'].fillna('')
df['text'] = df['text'].fillna('')
df['combined_text'] = (df['title'] + ' ' + df['text']).str.strip()

print('\nEmpty combined_text rows:', (df['combined_text'].str.len() == 0).sum())


for row in df['combined_text']:
    tokenized_str = row.split()


#map counts frequency of a token and the token itself, combining
counter_map = map(lambda x: (x, tokenized_str.count(x)), set(tokenized_str))

#sort the map by frequency
sorted_counter_map = sorted(counter_map, key=lambda x: x[1], reverse=True)

print('\nTop 10 most common tokens:')
for token, count in sorted_counter_map[:10]:
    print(f'{token}: {count}')

```
